```python
# import and load the data:

import pandas as pd
url = 'http://www.ldeo.columbia.edu/~rpa/usgs_earthquakes_2014.csv'
df = pd.read_csv(url, parse_dates=['time'], index_col='time')

df.head()
```

```python
# create a new dataframe with the 50 largest earthquakes
dfmax50 = df.nlargest(50, 'mag')

dfmax50.head()
```

```python
ccrs.Robinson?
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt


plt.figure(figsize=(10,7))
ax = plt.axes(projection=ccrs.Robinson(central_longitude=180))
ax.stock_img()


plt.scatter(dfmax50.longitude, dfmax50.latitude,c=dfmax50.mag,
            transform=ccrs.PlateCarree(), cmap='Reds', edgecolor='k')

plt.colorbar(shrink=.35, label='magnitude')
plt.title('Top 50 earthquakes of 2014')
```

```python
plt.figure(figsize=(10,7))
ax = plt.axes(projection=ccrs.Robinson(central_longitude=180))
ax.stock_img()


sc = plt.scatter(dfmax50.longitude, dfmax50.latitude,c=dfmax50.mag, s=2**dfmax50.mag,
            transform=ccrs.PlateCarree(), cmap='Reds', edgecolor='k')


# handles, labels = sc.legend_elements(prop="sizes", alpha=0.6)
# legend2 = ax.legend(handles, labels, loc="upper right", title="Sizes")


# plt.colorbar(shrink=.35, label='magnitude')
plt.title('Top 50 earthquakes of 2014')
```
