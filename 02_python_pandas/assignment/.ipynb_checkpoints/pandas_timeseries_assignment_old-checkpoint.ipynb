### 2c

Just like in class it looks like there are lots of 999 and 99.0 data. These probably aren't real, and if we look at the documentation for the buoy we will see that these are place-holders meaning 'bad data'. 

Again, reload the data, this time telling pandas that 999 and 99.0 are bad data and should be replaced with `NaN`. 

If you can't remember the option to tell pandas which values should be `NaN`, look back at the class notebook. Also google 'pandas read_csv'. Google is your friend. This should bring you to the pandas documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html . Here you will find a lot of info about all the options you can use with `pd.read_csv()`. You will find you use the documents for different libraries all the time. I do. 

As always, show how it's worked by printing the first 10 rows of the dataframe

````python
## answer here

df = pd.read_csv('44066h2017.txt', sep='\s+', na_values=[999, 99.0])

df.head()
````

### 2d

Oh man! We are almost there. What's the last thing we want to do? That's right, make a datetime index. We can see that we have columns for `#YY`, `MM`, `DD`, `hh` which are year, month, day, hour. We should be able to turn these into a pandas datetime object, then set the index of the dataframe to that. 

In class we used the `parse_dates= [column]` option in pandas to read in a column as a special datetime object. Now, we have to add a bunch of columns together to get a full representation of time. We can do this! We can actually do all the previous steps at once with a single line of code. Kinda makes sense right? to load and get out dataframe into shape we are doing to use the single line 
```python 
df = pd.read_csv('44066h2017.txt', sep='\s+', na_values=[999, 99.0], parse_dates={'date':[_, _, _, _]})
```

What `parse_dates={'date':[_, _, _, _]}` does is try to read the columns we give it as dates, then it squashes them all into one new column called 'date'. 'date' is just a name we are giving it. We could use 'banana' instead, but 'date' makes a lot of sense.

The only thing you need to do is fill in the blanks in parse dates. Remember, we are trying to use all the date columns together to get pandas to create the right time information. What should we do? Fill in the blanks and output the first 10 lines of the dataframe:

````python
# answer here

# df = pd.read_csv('44066h2017.txt', sep='\s+', na_values=[999, 99.0], parse_dates={'date':[0, 1, 2, 3]})
df = pd.read_csv('44066h2017.txt', sep='\s+', na_values=[999, 99.0], parse_dates=[[0, 1, 2, 3]])


df.head()
````
