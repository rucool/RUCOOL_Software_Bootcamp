Using the glob.glob() the filenames are listed and sorted.
From there I wrote a forloop that plots the standard deviation over time for all files listed.

```python
import glob
import numpy
import matplotlib.pyplot
```

```python
filename=sorted(glob.glob('../data/inflammation*.csv'))
```

```python
filename
```

```python
for file in filename:
    data = numpy.loadtxt(file,delimiter=',')
    std_data = numpy.std(data)
    matplotlib.pyplot.plot(numpy.std(data, axis=0))
    matplotlib.pyplot.xlabel('Time (days)')
    matplotlib.pyplot.ylabel('standard deviation of inflammation')
    matplotlib.pyplot.show()
    print(std_data)
    print(file)
```
