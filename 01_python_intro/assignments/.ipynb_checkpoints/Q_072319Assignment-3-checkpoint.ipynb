```python
# Full Code
import numpy 
import matplotlib.pyplot
import glob

InfFiles = sorted(glob.glob('../data/inflammation*.csv'))

for file in InfFiles:
    data = numpy.loadtxt(file, delimiter=',')
    std_data=numpy.std(data, axis=0)   
    Title = 'Standard Deviation Amoung Patients Per Day'
    
    matplotlib.pyplot.plot(std_data)
    matplotlib.pyplot.xlabel('Time (Days)')
    matplotlib.pyplot.ylabel('Standard Deviation')
    matplotlib.pyplot.title(Title)
    matplotlib.pyplot.figtext(.2, -.05, 'Data from file at pathname: ' + file)
    matplotlib.pyplot.show()
  

```
