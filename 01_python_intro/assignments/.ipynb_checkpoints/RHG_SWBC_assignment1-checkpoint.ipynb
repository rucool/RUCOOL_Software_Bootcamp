```python
# Import libraries
import numpy as np
import glob
import matplotlib.pyplot as plt

# Read in inflammation datasets
filenames = sorted(glob.glob('../data/inflammation*.csv'))
# Compute the standard deviation of each inflammation dataset
sd_vec = []
for file in filenames:
    data = np.loadtxt(file,delimiter=',')
    sd_data = np.std(data)
    sd_vec = np.append(sd_vec, sd_data)
print(sd_vec) # Check contents of sd_vec

#Plot array of standard deviations
fig = plt.figure(figsize=(10,5))
plt.plot(sd_vec,'go--')
plt.xlabel('# Dataset',fontsize=12)
plt.ylabel('Standard Deviation', fontsize=12)
```

A few things to note about the code above:
1) The answer array (sd_vec) needs to be initialized outside of the for loop before answers can be stored within it.
2) Text properties for the axis labels can be defined within the 'xlabel' and 'ylabel' functions of matplotlib.
3) Color and linestyle can be defined within the 'plot' function of matplotlib. For more elaborate colors, must use a number system.
