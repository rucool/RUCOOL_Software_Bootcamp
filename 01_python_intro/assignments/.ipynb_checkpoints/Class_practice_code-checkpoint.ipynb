```python
import numpy
import glob
import matplotlib.pyplot as plt

filename = sorted(glob.glob('../data/inflammation*.csv'))
print(filename)

for file in filename:
    data = numpy.loadtxt(file, delimiter = ',')
    mean_data = numpy.mean(data)
    #the zero will average over the column, 1 will average across the row
    plt.plot(numpy.mean(data, axis = 0))
    plt.xlabel('time')
    plt.ylabel('Mean Inflammation')
    
    print(mean_data)
 
```

```python
for file in filename:
    data = numpy.loadtxt(file, delimiter = ',')
    std_data = numpy.std(data)
    plt.plot(numpy.std(data, axis = 0))
    plt.xlabel('time')
    plt.ylabel('Std Inflammation')
    
    print(std_data)
    print(file)
```
