```python
# answer here
# by default if you call 'plt.plot()' more than once in a cell, the new stuff
# will be added to the same figure:

import matplotlib.pyplot as plt
import glob

# get the list of the data files
filenames = sorted(glob.glob('../data/inflammation*.csv'))

#create a 'for' loop that prints the filename, runs analyze():
for f in filenames:
    # print the current file in the loop:
    print(f)
    
    # load in the data
    data = numpy.loadtxt(f, delimiter=',')
    
    # add it to the plot, use the numpy.mean() function to get the mean
    plt.plot(numpy.mean(data, axis=0))
```

```python
# answer here

# You can speficy two input parameters that you can work with inside the function
# in this case I ask for feet and inches
# i also made a docstring (using '''''') for a description of the function

def feet2meters(f,i):
    '''take feed (f) and inches (i) and return meters''' 
    
    height = f + i/12
    height_m = height*0.3
    print('height is :', height_m)
    return height_m
```

```python
# load ARGO data file and examine keys
data = numpy.load('argo_float_4901412.npz')
list(data.keys())

```

```python
# extract the data into named variable arrays 
# print the shapes of a T, S, P and date, lat, lon, level

# I'm hoping that everyone sees that the data (temp, salinity, pressure)
# has a 2D shape: 78 x 75
# while the other things like lon, lat, date, levels are 1D
# and have shapes that match one of the dimension of the T and S data
# the lat, lon, levels, etc are basically the dimension labels , or coordinates of the data


# pull out the 'values', or the data, from the dictionary, using the keys
# we use the ['keyname'] to grab the data from a dict
T = data['T']
S = data['S']
P = data['P']
levels = data['levels']
date = data['date']

lon = data['lon']
lat = data['lat']


# you could print out the shape one at a time:
print('T shape is ', T.shape)
print('levels shape is ', levels.shape)

# or you could write a loop like this:
# inside the loop, k will change to be each of the keys
for k in list(data.keys()):
    print(k,' shape is ' , data[k].shape)
```

```python
# plot the first column of T, S, P vs levels
# I made two figures, but you could have done this with subplots

# create a figure
plt.figure()

# plot just the first column of T using the [] and indexing
plt.plot(T[:,0], levels)

# you can reverse the axis direction by reverseing the axis limits in 'ylim'
plt.ylim([80, 0])
plt.xlabel('temp')
plt.ylabel('level')


# create a second figure
plt.figure()


plt.plot(S[:,0], levels)
plt.ylim([80, 0])
plt.xlabel('salinity')
plt.ylabel('level')
```

```python
# use pcolormesh to plot the series of T and S profiles as a color panel
# flip the yaxis direction the same way as you did above
# add a colorbar

# here it was important to note that T and S have a shape that is like levels x date
# so a pcolorplot can be made with something like plt.pcolormesh( xdata, ydata, colordata)
# where xdata and ydata are 1D and match the 2D shape of colordata


plt.figure()
plt.pcolormesh(date, levels, T)
plt.ylim([80, 0])
plt.colorbar()

plt.figure()
plt.pcolormesh(date, levels, S)
plt.ylim([80, 0])
plt.colorbar()
```

```python
# get the mean of T and S at each depth level

meanT = numpy.nanmean(T, axis=1)
meanS = numpy.nanmean(S, axis=1)
```

```python
# plot the mean temp and salinity vs level (depth)
plt.figure()
plt.plot(meanT, levels)
plt.ylim([80, 0])
plt.xlabel('mean temperature')
plt.ylabel('level')

plt.figure()
plt.plot(meanS, levels)
plt.ylim([80, 0])
plt.xlabel('mean salinity')
plt.ylabel('level')


# the errorbar function is something you can look up by googling 'matplotlib errorbar'
plt.figure()
plt.errorbar(meanS, levels, xerr = numpy.nanstd(S,axis=1))
         
plt.ylim([80, 0])
plt.xlabel('mean salinity')
plt.ylabel('level')
```

```python
# make a scatter plot of the lat and lon

plt.figure()
plt.scatter(lon, lat)
plt.grid()

# you can also color the dots based on data of the same size:
# like here color by date
plt.figure()
plt.scatter(lon, lat, c=date)
plt.grid()


# a more meaninful plot might have some temperature info
# for temp to be the same size as lat and lon, and to work
# in the plot, we need to remove the 'depth', or levels dimention
# in otherwords we need to average or something over depth at each lat & lon

# we can plot a scatter of the depth-averaged temperature like this:

plt.figure()
plt.scatter(lon, lat, c=T.mean(axis=0))
plt.grid()
# i can also add a label and units to the colorbar
# $^\circ$ is a way to make a degree sign
plt.colorbar(label='depth-avg Temp [$^\circ$C]')
```
