How do you concatenate a numpy array in a python for loop if we wanted to look at the standard deviation of a whole data file compared to the other files?

```python
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
fnames = sorted(glob.glob('../data/*inflam*')
print(fnames)
```

```python
for file in fnames:
    data = np.loadtxt(file, delimiter=',')
    
    plt.plot(np.std(data, axis = 0))
    plt.xlabel('Days')
    plt.ylabel('Standard deviation of inflammation')
    plt.title(file)
    plt.show()
```
