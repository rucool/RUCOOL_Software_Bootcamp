```python
import numpy
import glob
import matplotlib.pyplot
```

```python
HW1 = glob.glob('../data/inflammation*.csv')
print(HW1)
```

```python
# loop through data and calc mean
import matplotlib.pyplot

for file in HW1:
    data = numpy.loadtxt(file,delimiter=',')
    std_data = numpy.std(data)
    
    matplotlib.pyplot.plot(numpy.std(data, axis=0))
    matplotlib.pyplot.show()
    matplotlib.pyplot.title('Standard Deviation')
    matplotlib.pyplot.xlabel('Days')
    matplotlib.pyplot.ylabel('Patient #')
    
    print(std_data)
    print(file)
```
