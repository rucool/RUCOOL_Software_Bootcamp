```python
import numpy
```

```python
import glob
```

```python
import matplotlib.pyplot
```

```python
glob.glob('../data/inflammation*.csv')
```

```python
filenames = glob.glob('../data/inflammation*.csv')
```

```python
for file in filenames:
    data = numpy.loadtxt(file,delimiter=',')
    
    matplotlib.pyplot.plot(numpy.std(data,axis=0))
    matplotlib.pyplot.xlabel('Days')
    matplotlib.pyplot.ylabel('Standard Deviation of Inflammation')
    matplotlib.pyplot.show()
    
    stdev_data = numpy.std(data)
    print('Overall Standard Deviation of',file,'is',stdev_data)
  
```

This code ran smoothly, like the similar example we did in class. It's interesting that the standard deviation of inflammation follows a pattern similar to the inflammation data itself - higher average inflammation generally has higher standard deviation. 
The hardest part about this was remembering to include all parts necessary in the commands, e.g. I would tend to forget the "axis=0" part of numpy.std(data,axis=0) or the "plot" part of matplotlib.pyplot.plot(...). I think that will get easier with more practice.
